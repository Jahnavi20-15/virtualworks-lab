```python
import pandas as pd

df = pd.read_excel("company_dataset.csv.xlsx")

print(df.head())
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
# Fill missing salary with average salary
df["Salary"].fillna(df["Salary"].mean(), inplace=True)

# Fill missing names
df["Name"].fillna("Unknown", inplace=True)
```

```python
duplicates = df[df.duplicated()]
print(duplicates)

# Remove duplicates
df = df.drop_duplicates()
```

```python
df["Department"] = df["Department"].replace({
    "hr": "HR",
    "Hr": "HR",
    "H.R.": "HR",
    "sales ": "Sales",
    "Information Tech": "IT"
})
```

```python
df["Joining_Date"] = pd.to_datetime(
    df["Joining_Date"],
    errors="coerce"
)

print(df["Joining_Date"])
```

```python
print(df.dtypes)
```

```python
df["Salary"] = pd.to_numeric(df["Salary"], errors="coerce")
```

```python
# Remove negative salary
df = df[df["Salary"] > 0]

# Keep scores between 1 and 5
df = df[df["Performance_Score"].between(1,5)]
```

```python
print(df["Employee_ID"].is_unique)
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
