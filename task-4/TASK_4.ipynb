```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {
    'User_ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Page_Visited': ['Home', 'Products', 'Pricing', 'Blog', 'Products',
                     'Contact', 'Home', 'Pricing', 'Products', 'Contact'],
    'Session_Duration_Min': [5, 12, 18, 3, 10, 2, 6, 20, 15, 1],
    'Bounce': [0, 0, 0, 1, 0, 1, 0, 0, 0, 1],
    'Exit_Page': ['Products', 'Pricing', 'Contact', 'Blog', 'Products',
                  'Contact', 'Home', 'Pricing', 'Products', 'Contact']
}
df = pd.DataFrame(data)
print("WEBSITE TRAFFIC DATASET")
print(df)
print("\nDataset Info:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())
page_views = df['Page_Visited'].value_counts()
print("\nMost Visited Pages:")
print(page_views)
plt.figure(figsize=(8,5))
page_views.plot(kind='bar', color='skyblue')
plt.title("Most Visited Pages")
plt.xlabel("Pages")
plt.ylabel("Views")
plt.show()
avg_duration = df['Session_Duration_Min'].mean()
print(f"\nAverage Session Duration: {avg_duration:.2f} minutes")
bounce_rate = df['Bounce'].mean() * 100
print(f"\nBounce Rate: {bounce_rate:.2f}%")
exit_analysis = df['Exit_Page'].value_counts()
print("\nExit Page Analysis:")
print(exit_analysis)
plt.figure(figsize=(8,5))
exit_analysis.plot(kind='pie', autopct='%1.1f%%')
plt.title("Exit Page Distribution")
plt.ylabel("")
plt.show()
print("\nWebsite Improvement Suggestions:")
print("1. Improve Contact page because many users exit there")
print("2. Optimize Blog content to reduce bounce rate")
print("3. Enhance Pricing page for better conversions")
print("4. Improve page loading speed and navigation")
df.to_csv("web_traffic_analysis.csv", index=False)
print("\nAnalysis saved as web_traffic_analysis.csv")
```
